### MACD Indikatörü

**MACD**, Moving Average Convergence/Divergence Hareketli ortalamaların birbiri ile yakınlaşması veya uzaklaşması ile elde edilen finans piyasalarının teknik analizinde kullanılan son derece popüler bir trend göstergesidir. MACD, bir hisse senedi, kripto paralar veya başka bir ticaret elde edilebilir finansal varlığın momentumunu belirlemek için hareketli ortalamaları kullanan ve trendi takip eden bir araçtır.

MACD, öncü trend yönü ve süresini saptamak üzere iki farklı uzunluğa sahip hareketli ortalama değerini kullanmaktadır. Uzun vadeli olarak 26 gözlem, kısa vadeli olarak da 12 gözlem sayısına sahip üssel hareketli ortalama en çok kullanılan değerdir. Genellikle 26 günlük üssel hareketli ortalama ve 12 günlük üssel hareketli ortalama değeri kullanılır.

**Bu indikatörün** kullanıldığı grafiklerde iki hareketli ortalamanın EMA değerleri (Üssel Hareketli Ortalama) Sinyal Çizgisi ve İki hareketli ortalamanın arasındaki fark **Gösterge** çizgisini oluşturur. Ortada sıfır çizgisi altında ve üstünde hareket eden bir histogram grafiği oluşur.

#### MACD Indikatörü Nasıl Çalışır ?

MACD göstergesi sıfır çizgisinin etrafında hareket eden üç elementten oluşur:

**MACD Hattı (1):** Yukarı veya aşağı doğru hareket eden momentumun (piyasa trendinin) belirlenmesine yardımcı olur. İki üstel hareketli ortalamalayı (EMA) çıkararak hesaplanır.

**Sinyal Hattı (2):** MACD hattının bir üstel hareketli ortalamaları (EMA) (genellikle 9 dönem EMA). MACD hattı ile sinyal hattının kombine analizi ile alınacak kararı potansiyel ters yöne çevirmede veya giriş ve çıkış noktaları tespitinde yardımcı olabilir.

**Histogram (3):** MACD hattının ve sinyal hattının uzaklaşma ve yakınsamasının grafiksel bir gösterimidir. Başka bir deyişle, histogram iki satır arasındaki farklılıklara göre hesaplanır.

#### MACD Hattı (1):

Genel olarak üstel hareketli ortalamalar, bir varlığın kapanış fiyatlarına göre ölçülür ve iki dönemi hesaplamak için kullanılan dönemler genellikle 12 dönem (daha hızlı) ve 26 dönem (daha yavaş) olarak ayarlanır. Dönemler farklı şekillerde (dakika, saat, gün, hafta, ay) yapılandırılabilir, ancak bu makale günlük ayarlara odaklanacaktır. Yine de, MACD göstergesi farklı ticaret stratejilerini karşılamak üzere özelleştirilebilir.

Standart zaman aralıklarını varsayarsak, MACD hattının kendisi, 26 günlük EMA’yı 12 günlük EMA’dan çıkararak hesaplanır:

MACD hattı = 26 gün EMA - 12 EMA


#### Sinyal Hattı (2):

Varsayılan olarak, sinyal hattı ana hattın 9 günlük bir EMA’sından hesaplanır ve bu nedenle önceki hareketleri hakkında daha fazla bilgi sağlar. Her zaman doğru olmasa da, MACD hattı ve sinyal hattı kesiştiğinde, özellikle MACD grafiğinin ekstremitelerinde (sıfır çizgisinin çok üstünde veya çok altında) meydana geldiklerinde, bu olaylar genellikle eğilim ters sinyalleri olarak kabul edilir.

Sinyal hattı = MACD Hattı'nın 9 günlük EMA’sı

#### MACD Histogramı (3):

Histogram, MACD hattının ve sinyal hattının göreli hareketlerinin görsel bir kaydından başka bir şey değildir. Sadece birinden diğerine çıkarılarak hesaplanır. Bununla birlikte, üçüncü bir hareketli çizgi eklemek yerine, histogram bir çubuk grafiğinden yapılır ve görsel olarak okumayı ve yorumlamayı kolaylaştır.

MACD histogramı = MACD Hattı - Sinyal Hattı

#### MACD Indikatör Ayarları Nasıl Yapılır ?

MACD için varsayılan ayarlar 12, 26, 9 gün aralığına dayanır. Fakat bazı analistler daha hassas bir gösterge oluşturmak için günleri 5, 35, 5 gün olacak şekilde de ayarlayabilmektedir. Kripto para piyasalarının yüksek volatilitesinden dolayı, MACD göstergesinin duyarlılığının artması riskli olabilir, çünkü büyük olasılıkla daha yanlış sinyaller ve yanıltıcı bilgiler doğuracaktır.

#### MACD Indikatörü Nasıl Yorumlanır ?

MACD göstergesinin en önemli ve en çok kullanılan 3 tane yorumlanma methodu vardır:

**Kesişmeler:** MACD göstergesi sinyal çizgisinin altına düşerse bu aşağı yönlü bir sinyal oluşturmakta olup satış zamanının geldiğine işaret edebilmektedir. Tersi durumda ise MACD göstergesi sinyal çizgisinin üzerine çıkarsa bu yukarı yönlü bir sinyal oluşturmakta ve alış seviyesinin geldiğine işaret edebilmektedir. Çoğu yatırımcı, yanlış sinyallerin önüne geçebilmek için kesişme seviyesinin de üzerini beklemektedir.

**Ayrışmalar:** Finansal bir ürünün fiyatı MACD göstergesinden ayrışırsa bu mevcut trendin sona erdiği anlamına gelmektedir.

**Anormal Yükselişler:** MACD göstergesi anormal bir şekilde yükselirse bu kısa vadeli hareketli ortalamanın uzun vadeli hareketli ortalamayı yukarı yönlü ittiğini göstermektedir. Bu da ilgili finansal ürünün aşırı alım bölgesinde olduğu ve kısa sürede normal haline döneceğine işaret etmektedir. Yatırımcılar ayrıca sıfır çizgisinin altını ve üstünü de sinyal olarak dikkate almaktadırlar. MACD göstergesi sıfır çizgisinin üzerindeyken, kısa vadeli ortalama uzun vadeli ortalamanın üzerindedir ve yukarı yönlü hareketin devam edebileceğini göstermektedir. Sıfır çizgisinin altı ise bunun tam tersini ifade etmektedir.

```python
# Gerekli kütüphane ve fonksiyonların kurulması

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm

import yfinance as yf
import plotly.graph_objects as go

import warnings
warnings.filterwarnings("ignore")
```

**Veri Okuma İşlemi | Yahoo Finance**

Önceki yazılarımızda Yahoo Finance üzerinden veri çekmemizi sağlayan bir fonksiyon yazmıştık. Onun yardımıyla bir hisse senedinin verilerine erişelim.

```python
def dataImporter(symbol = "", start_date = '2017-01-01', inBist = True):
    if inBist:
        symbol = symbol + ".IS"
        df = yf.download(symbol,
                         start = start_date,
                         progress = False)

    else:
        df = yf.download(symbol,
                         start = start_date,
                         progress = False)
    return df
```

AMAZON hisse senedine ait verileri **df** adında bir dataframe’de tutalım:

```python
df = dataImporter("AMZN", inBist = False)

df.head()
```

Hisse senedine ait kapanış değerlerini görselleştirelim:

```python
df["Close"].plot()
```

Şimdi adım adım MACD’yi hesaplayalım.

İlk olarak hesaplamamız gereken değer MACD Hattı değeri. Bunu da 26 günlük EMA’dan 12 günlük EMA çıkararak buluyorduk.

12 Günlük EMA’nın hesaplanması:

```python
df["Short"] = df['Close'].ewm(span = 12, adjust = False).mean()
```

26 Günlük EMA’nın hesaplanması:

```python
df["Long"] = df['Close'].ewm(span = 26, adjust = False).mean()
```

MACD Hattı’nın hesaplanması:

```python
df["MACD"] = df["Short"] - df["Long"]
```

MACD Hattı’nın görselleştirilmesi:

```python
df["MACD"].plot()
```

MACD Hattı’nın hesaplanmasının ardından Sinyal Hattı’nın hesaplanması gerekmektedir:

```python
df['Signal'] = df['MACD'].ewm(span = 9, adjust = False).mean()
```

Ardından Sinyal değerlerimiz ile MACD değerlerimizi görsel üzerinde inceleyelim. Signal ve MACD değerlerin kesişim noktaları bizlere AL/SAT tavsiyelerinde yardımcı yorumlar verebilir.

```python
plt.figure(figsize = (10, 4))
plt.plot(df.index, df.MACD, color = "red")
plt.plot(df.index, df.Signal, color = "blue")
plt.show()
```
